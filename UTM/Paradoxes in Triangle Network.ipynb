### For (a):
$\min Z(x)=0.5x_a^2+x_a + 0.5x_b^2+4x_b + 0.5x_c^2+x_c$ \
s.t. \
$f_{1-2}=1$ \
$f_{1-2-3}+f_{1-3}=2$ \
$x_a=f_{1-2}+f_{1-2-3}$ \
$x_b=f_{1-3}$ \
$x_c=f_{1-2-3}$ \
$f_{1-2-3}, f_{1-3},f_{1-2} \geq 0$ \
$x_a,x_b,x_c \geq 0$

```python
def objective(x):
    x1 = x[0] #x_a
    x2 = x[1] #x_b
    x3 = x[2] #x_c
    x4 = x[3] #f_{1-2}
    x5 = x[4] #f_{1-2-3}
    x6 = x[5] #f_{1-3}
    return 0.5*x[0]**2+x[0] + 4*x[1]+0.5*x[1]**2 + x[2]+0.5*x[2]**2 

def constraint1(x):
    return x[3]-1

def constraint2(x):
    return x[4]+x[5]-2

def constraint3(x):
    return x[0]-x[3]-x[4]

def constraint4(x):
    return x[1]-x[5]

def constraint5(x):
    return x[2]-x[4]

def equ_time_a(x): 
    return x[0]+1

def equ_time_b(x): 
    return x[1]+4

def emi(x):
    return 0.01*x[1]+0.01*x[0]+0.5*x[2]
```

```python
bnds = [(0, None), (0, None), (0, None), (0, None), (0, None), (0, None)]
cons = [{'type': 'eq', 'fun': constraint1},
        {'type': 'eq', 'fun': constraint2},
        {'type': 'eq', 'fun': constraint3},
        {'type': 'eq', 'fun': constraint4},
        {'type': 'eq', 'fun': constraint5}]
x_ini = [0,0,0,0,0,0]
```

```python
res = minimize(objective, x_ini, constraints=cons, bounds=bnds, tol=0.0001)
res.x
```

```python
res.fun
```

```python
print('$x_{a}^*=',res.x[0])
print('$x_{b}^*=',res.x[1])
print('$x_{c}^*=',res.x[2])
print("Equilibrium paths time for (1,2): ", equ_time_a(res.x))
print("Equilibrium paths time for (1,3): ", equ_time_b(res.x))
print("Total emissions: ", emi(res.x))
```
