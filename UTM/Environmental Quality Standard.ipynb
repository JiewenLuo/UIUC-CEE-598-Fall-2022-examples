```python
from IPython.display import Image
import numpy as np
from scipy.optimize import minimize
Image(filename = './Environmental Quality Standard.jpg')
```

```python
def objective(x):
    x1 = x[0] #x_a
    x2 = x[1] #x_b
    x3 = x[2] #x_c
    x4 = x[3] #Q
    return x[3]

def constraint1(x):
    return x[0]+x[1]+x[2]-10

def constraint2(x):
    return x[3]-(0.1*x[0]+0.2*x[1]+0.3*x[2])
```

```python
bnds = [(0, None), (0, None), (0, None), (0, None)]
cons = [{'type': 'eq', 'fun': constraint1},
        {'type': 'ineq', 'fun': constraint2}]
x_ini = [0,0,0,0]
```

```python
res = minimize(objective, x_ini, constraints=cons, bounds=bnds, tol=0.0001)
res.x
```

```python
print('Q*=',res.fun)
```
