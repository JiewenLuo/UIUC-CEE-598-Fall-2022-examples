```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize
plt.style.use('ggplot')
```

```python
a = 1
b = 3

x = np.linspace(a, b)

def f(x):
    return 65*x**6+71*x**5-322*x**4-401*x**3
def df(x):
    return 65*6*np.power(x,5)+71*5*np.power(x,4)-322*4*np.power(x,3)-401*3*np.power(x,2)

plt.plot(x, f(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```

```python
c=0.618*a+(1-0.618)*b
d=(1-0.618)*a+0.618*b
if f(c)<f(d):
    b=d
else:
    a=c
print(c,d,a, b)
x = np.linspace(a, b)
plt.plot(x, f(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```
