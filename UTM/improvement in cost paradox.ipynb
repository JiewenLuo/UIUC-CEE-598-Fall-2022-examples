### For the original one:
$\min Z(x)=0.5x_a^2 +5x_a+ 0.5x_b^2+9x_b$ \
s.t. \
$f_{a}+f_b=6$ \
$x_a=f_{a}$ \
$x_b=f_{b}$ \
$f_{a}, f_{b} \geq 0$ \
$x_a,x_b \geq 0$

```python
def objective(x):
    x1 = x[0] #x_a
    x2 = x[1] #x_b
   
    return 0.5*x[0]**2 + 5*x[0]+0.5*x[1]**2+ 9*x[1]

def constraint1(x):
    return x[0]+x[1]-6

def equ_time(x): 
    return 5+x[0]

def emi(x):
    return 0.4*x[1]+x[0]*0.5
```

```python
bnds = [(0, None), (0, None)]
cons = [{'type': 'eq', 'fun': constraint1}]
x_ini = [0,0]
```

```python
res = minimize(objective, x_ini, constraints=cons, bounds=bnds, tol=0.0001)
res.x
```

```python
print('$x_{a}^*=',res.x[0])
print('$x_{b}^*=',res.x[1])
print("Equilibrium paths time: ", equ_time(res.x))
print("Total emissions: ", emi(res.x))
```
